```python
from sklearn.decomposition import PCA
```

```python
p = PCA(n_components = 50 )
```

```python
p
```

```python
p1=p.fit_transform(i_a1)
```

```python
p1
```

```python
import matplotlib.pyplot as plt
```

```python
p2 = p.inverse_transform(p1)
```
